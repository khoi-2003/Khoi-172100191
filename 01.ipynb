```python
# Import thư viện
from PIL import Image
```

```python
# Thư mục chứa ảnh
# Trong python dùng dấu / hoặc \\
import os
print(os.getcwd()) #In thư mục hiện tại
directory = 'C:/Users/Admin/Documents/GTA San Andreas User Files/SAMP/screens'
```

```python
#Tạo đường dẫn đến ảnh
image_path = directory + '/sa-mp-000.png'
print (image_path)
```

```python
#Sử dụng Image.open() đọc ảnh
img01 = Image.open(image_path)
```

```python
# Hiển thị hình ảnh
display(img01)
```

```python
# In một số thông tin
# In định dạng ảnh
print("Định dang ảnh: ", img01.format)
```

```python
# In ra size
print("Kích thước hình ảnh: ", img01.size)
```
