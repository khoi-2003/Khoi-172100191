```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

# Đọc ảnh
image_path = 'C:/Users/Admin/Documents/GTA San Andreas User Files/SAMP/screens/sa-mp-006.png'
image = Image.open(image_path)
```

```python
# Ảnh xám
gray_image = image.convert('L')

# Chuyển ảnh xám về mảng
image_a = np.array(gray_image)

# Tính gradient theo hướng x
gradient_x = ndimage.sobel(image_a, axis=0, mode='constant')

# Hiển thị ảnh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('gray_image')

plt.subplot(2,1,2)
plt.imshow(gradient_x, cmap='gray')
plt.title('gradient_x')

plt.show()
```
